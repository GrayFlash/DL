```python
from PIL import Image, ImageEnhance
import pytesseract as pt
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('bill1.jpeg',0)
```

```python
# Pre-processing

img = Image.fromarray(np.array(img))
contrast = ImageEnhance.Contrast(img)
img = contrast.enhance(2)
img_array = np.array(img)
blurred = cv.GaussianBlur(img_array, (5,5), 0)
threshold = cv.adaptiveThreshold(blurred, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 15, 5)

```

```python
img_thresh_rgb = cv.cvtColor(threshold, cv.COLOR_BGR2RGB)
plt.imshow(img_thresh_rgb)
```

```python
print(pt.image_to_string(img_thresh_rgb))
```
